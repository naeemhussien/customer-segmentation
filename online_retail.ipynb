```python
# df3.groupby(['CustomerID']).first()
dfcon = df3.groupby('CustomerID').agg({'InvoiceNo': 'count',
                                       'Total Price' : 'sum', 
                                       'InvoiceDate' : lambda x: (datenow - x.max()).days})
dfcon
```

```python
dfcon
```

```python
def RScore(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
    
def FMScore(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1
```

```python
quantiles = dfcon.quantile(q=[0.25,0.5,0.75])
```

```python
dfcon['r_quartile'] = dfcon['InvoiceDate'].apply(RScore, args=('InvoiceDate',quantiles)).map(str)
dfcon['f_quartile'] = dfcon['InvoiceNo'].apply(FMScore, args=('InvoiceNo',quantiles)).map(str)
dfcon['m_quartile'] = dfcon['Total Price'].apply(FMScore, args=('Total Price',quantiles)).map(str)
```

```python
dfcon['rfm'] = dfcon['r_quartile'] + dfcon['f_quartile'] + dfcon['m_quartile']
dfcon
```

```python
dfcon.sort_values(['rfm'], ascending= True)
```

```python
def RtestScore(x,p,d):
    if x <= d[p][0.2]:
        return 1
    elif x <= d[p][0.4]:
        return 2
    elif x <= d[p][0.6]: 
        return 3
    elif x <= d[p][0.8]:
        return 4
    else:
        return 5
    
quantiles1 = dfcon.quantile(q=[0.2,0.4,0.6,0.8])
```

```python
dfcon['rtest_quartile'] = dfcon['InvoiceDate'].apply(RtestScore, args=('InvoiceDate',quantiles1)).map(str)
dfcon
```
